# Exploratory Data Analysis & K-Means Clustering

Notebook ini digunakan untuk mendemonstrasikan metodologi (25% Bobot Penilaian) yang meliputi eksplorasi data, _data cleaning_, dan pembuatan model _K-Means Clustering_ berdasarkan data ekonomi regional.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
df = pd.read_csv('../data/raw/bps_economic_data.csv')
df.head()
```

## Visualisasi Distribusi PDRB dan QRIS

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Volume_QRIS_Juta', y='PDRB_Triliun', size='Tingkat_Kemiskinan_Persen', sizes=(20, 200))
plt.title('Hubungan Transaksi Digital vs PDRB')
plt.show()
```
